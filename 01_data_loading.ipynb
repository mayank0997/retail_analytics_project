### Data Ingestion

Imports and Spark configuration

```python
# Data Loading Notebook
# Purpose: Load Instacart CSV files from Lakehouse into Bronze Delta tables

from pyspark.sql import functions as F

spark.conf.set("spark.sql.shuffle.partitions", "200")
```

### Define Lakehouse File Paths

Setting base path

```python
base_path = "Files/"
```

Defining each dataset

```python
orders_path = base_path + "orders.csv"
order_products_prior_path = base_path + "order_products__prior.csv"
order_products_train_path = base_path + "order_products__train.csv"
products_path = base_path + "products.csv"
aisles_path = base_path + "aisles.csv"
departments_path = base_path + "departments.csv"
```

### Read CSV Files into Spark

Header and schema are inferred.

```python
common_csv_options = {
    "header": "true",
    "inferSchema": "true",
    "quote": '"',
    "escape": '"',
    "multiLine": "true"
}
```

```python
orders = spark.read.format("csv") \
    .options(**common_csv_options) \
    .load(orders_path)

order_products_prior = spark.read.format("csv") \
    .options(**common_csv_options) \
    .load(order_products_prior_path)

order_products_train = spark.read.format("csv") \
    .options(**common_csv_options) \
    .load(order_products_train_path)

products = spark.read.format("csv") \
    .options(**common_csv_options) \
    .load(products_path)

aisles = spark.read.format("csv") \
    .options(**common_csv_options) \
    .load(aisles_path)

departments = spark.read.format("csv") \
    .options(**common_csv_options) \
    .load(departments_path)
```

### Quick Validation

```python
orders.printSchema()
orders.show(5)
order_products_prior.printSchema()
order_products_prior.show(5)
order_products_train.printSchema()
order_products_train.show(5)
departments.printSchema()
departments.show(5)
aisles.printSchema()
aisles.show(5)
products.printSchema()
products.show(5)
```

### Write Bronze Delta Tables

Writing the tables into the Lakehouse as Delta tables.

```python
orders.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("bronze_orders")

order_products_prior.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("bronze_order_products_prior")

order_products_train.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("bronze_order_products_train")

products.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("bronze_products")

aisles.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("bronze_aisles")

departments.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("bronze_departments")
```
